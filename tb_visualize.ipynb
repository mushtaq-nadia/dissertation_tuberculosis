```python
import matplotlib.pyplot as plt

class_counts = {}

# Iterate through subfolders (classes) in the dataset folder
for class_name in os.listdir():
    if os.path.isdir(class_name):
        class_count = len(os.listdir(class_name))
        class_counts[class_name] = class_count

# Plot the class counts
plt.figure(figsize=(6, 4))
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel('Classes')
plt.ylabel('Image Count')
plt.title('Image Counts per Class')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Assuming your dataset is mounted and the path is set
dataset_path = '/content/drive/MyDrive/tb'
os.chdir(dataset_path)

# Count the number of images in each class (before balancing)
class_counts = {}

for class_name in os.listdir():
    if os.path.isdir(class_name):
        class_count = len(os.listdir(class_name))
        class_counts[class_name] = class_count

# Calculate the total number of images before balancing
total_images_before = sum(class_counts.values())

# Assuming you balance the dataset and store the new class counts
balanced_class_counts = {
    'TB': 2100,
    'NON-TB': 2100

    # ...
}

# Calculate the total number of images after balancing
total_images_after = sum(balanced_class_counts.values())

# Plotting the class distribution before and after balancing
plt.figure(figsize=(10, 6))

# Before balancing
plt.bar(class_counts.keys(), class_counts.values(), color='red', alpha=0.5, label='Before Balancing')

# After balancing
plt.bar(balanced_class_counts.keys(), balanced_class_counts.values(), color='green', alpha=0.5, label='After Balancing')

plt.xlabel('Classes')
plt.ylabel('Image Count')
plt.title('Class Distribution Before and After Balancing')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# Print the total number of images before and after balancing
print(f"Total images before balancing: {total_images_before}")
print(f"Total images after balancing: {total_images_after}")
```
